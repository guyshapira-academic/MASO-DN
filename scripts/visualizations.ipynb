```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

```python
# 2D Sanity Check
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(10, 10))
plt.subplots_adjust(wspace=0, hspace=0)
for i, name in enumerate(["moons", "circles", "both"]):
    if i == 1:
        continue
    for j, ptype in enumerate(["local", "local", "global"]):
        if j == 2:
            img_name = f"images/{ptype}_{name}.png"
            img_title = r"$\omega_{g}^{(2)}$"
        else:
            img_name = f"images/{ptype}_{name}_{j}.png"
            if j == 0:
                img_title = r"$\omega_{in}^{(1)}$"
            else:
                img_title = r"$\omega_{in}^{(2)}$"
        img = Image.open(img_name)
        axs[i, j].imshow(img)
        axs[i, j].axis("off")
        axs[i, j].set_title(img_title, fontsize=20)
plt.tight_layout()
# plt.savefig("2d_sanity_check.png")
plt.show()
plt.close()
```

```python
# MASO CNN
name = "trained"
for i in range(10):
    fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(10, 10))
    plt.subplots_adjust(wspace=0, hspace=0)
    for j, layer_name in enumerate(["ReLU 1", "Max Pool 1", "ReLU 2", "Max Pool 2", "ReLU 3"]):
        img_name_trained = f"images/cnn_image_{i}_layer_{j+1}_{name}.png"
        img = Image.open(img_name_trained)
        axs[j].imshow(img)
        axs[j].axis("off")
        axs[j].set_title(layer_name, fontsize=14)
    plt.tight_layout()
    plt.savefig(f"images/image_{i}_{name}.png", bbox_inches='tight', pad_inches=0)
    plt.show()
    plt.close()
```
